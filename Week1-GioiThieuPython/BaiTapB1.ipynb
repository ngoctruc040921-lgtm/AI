```python
#15
class SinhVien:
    def __init__(self, ma, ten, namsinh, dtb):
        self.ma = ma
        self.ten = ten
        self.namsinh = namsinh
        self.dtb = dtb

# c. đủ điều kiện lên lớp
def dem_len_lop(ds):
    return sum(1 for sv in ds if sv.dtb >= 5)

# d. đủ 20 tuổi
def sv_20_tuoi(ds):
    return [sv for sv in ds if 2026 - sv.namsinh >= 20]

# e. hệ DH
def dem_DH(ds):
    return sum(1 for sv in ds if sv.ma[2:4] == "DH")

# f. tên Lan
def dem_ten_lan(ds):
    return sum(1 for sv in ds if "Lan" in sv.ten)

# g. họ Phan
def dem_ho_phan(ds):
    return sum(1 for sv in ds if sv.ten.startswith("Phan"))
```
