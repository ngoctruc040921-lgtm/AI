```python
#Bai1
def tron_mang(a, b):
    c = []
    n = min(len(a), len(b))

    # Lấy min từng cặp
    for i in range(n):
        c.append(min(a[i], b[i]))

    # Nếu còn phần dư thì cộng lại
    if len(a) > len(b):
        c.extend(a[n:])
    else:
        c.extend(b[n:])

    return c
```

```python
#Bai2
#a
def tong_tam_giac_tren_cheo_phu(a, n):
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                tong += a[i][j]
    return tong
```

```python
#b
def chuyen_am(a):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])
```

```python
#c
def thay_chan(a, x):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] % 2 == 0:
                a[i][j] = x
```

```python
#d
def toan_chan(a):
    for row in a:
        for val in row:
            if val % 2 != 0:
                return False
    return True
```

```python
#e
def doi_xung(a, n):
    for i in range(n):
        for j in range(n):
            if a[i][j] != a[j][i]:
                return False
    return True
```

```python
#f
def cheo_chinh_tang(a, n):
    for i in range(n - 1):
        if a[i][i] >= a[i+1][i+1]:
            return False
    return True
```

```python
#g
def xuat_tam_giac_duoi_cheo_phu(a, n):
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(a[i][j], end=" ")
        print()
```

```python
#h
def cheo_phu_giam(a, n):
    for i in range(n - 1):
        if a[i][n-1-i] <= a[i+1][n-2-i]:
            return False
    return True
```

```python
#Cau5
import pandas as pd
import matplotlib.pyplot as plt

def demo_pandas_matplotlib():
    data = {
        "Nam": [2020, 2021, 2022, 2023],
        "DoanhThu": [100, 150, 200, 250]
    }

    df = pd.DataFrame(data)
    print(df)

    plt.plot(df["Nam"], df["DoanhThu"])
    plt.xlabel("Năm")
    plt.ylabel("Doanh thu")
    plt.title("Biểu đồ")
    plt.show()
```
